```python
#Simulação de dois tanques, um superior com volume inicial;
#Um segundo vazio de inicio;

#Carregando bibliotecas
import scipy as sp
import numpy as np

#Dados principais
raio = 0.5
area = np.pi*(raio**2)
altura_h1 = 1.25
altura_h2 = 0
volume_1 = area*altura_h1
volume_2 = area*altura_h2
area_or_01 = 0.0003
area_or_02 = 0.0002
c = 0.5
grav = 9.81
Q1 = c*area_or_01*np.sqrt(2*grav*altura_h1)
Q2 = c*area_or_02*np.sqrt(2*grav*altura_h2)
T = 0
DT = 5

#Exibindo dados
print(f' {raio} {area:,.3f} {altura_h1} {volume_1:,.3f} {area_or_01} {area_or_02} {c} {grav} {Q1:,.5f} {Q2:,.5f}');

#Modelando
while (T < 3735):
    if(altura_h1 > 0.05):
        Q1 = c * area_or_01 * np.sqrt(2 * grav * altura_h1)
    else:
        Q1 = 0

    if(altura_h2 > 0.05):
        Q2 = c * area_or_02 * np.sqrt(2 * grav * altura_h2)
    else:
        Q2 = 0
    print(f'Tempo {T}, Altura h1 {altura_h1:,.3f}, Altura h2 {altura_h2:,.3f}, Volume 1 {volume_1:,.3f}, Volume 2 {volume_2:,.3f}, Vazão Q1 {Q1:,.5f}, Vazão Q2 {Q2:,.5f}.');
    VF1 = volume_1 - DT * Q1
    VF2 = volume_2 + DT * (Q1 - Q2)
    T = T + DT
    volume_1 = VF1
    volume_2 = VF2
    altura_h1 = volume_1 / area
    altura_h2 = volume_2 / area
```
